```python
import os
import shutil

def organize_downloads(source_dir):
    """
    Organizes files in the specified downloads folder based on their extensions.

    Args:
        downloads_folder (str): The path to the downloads folder.

    Raises:
        OSError: If an error occurs while creating folders or moving files.
    """

    # Define a dictionary to map file extensions to folder names (customize as needed)
    file_type_folders = {
        '.pdf': 'Documents/PDFs',
        '.docx': 'Documents/Word',
        '.xlsx': 'Documents/Excel',
        '.pptx': 'Documents/Presentations',
        '.jpg': 'Images',
        '.jpeg': 'Images',
        '.png': 'Images',
        '.gif': 'Images',
        '.mp3': 'Music',
        '.mp4': 'Videos',  # Update the folder name for videos
        '.zip': 'Compressed',
        '.rar': 'Compressed',
    }

    # Create the downloads folder path if it doesn't exist
    os.makedirs(source_dir, exist_ok=True)

    for filename in os.listdir(source_dir):
        # Get the full path of the file
        file_path = os.path.join(source_dir)

        if os.path.isfile(file_path):
            # Extract file extension (lowercase for case-insensitive matching)
            file_extension = os.path.splitext(filename)[1].lower()

            # Determine the destination folder based on the extension
            destination_folder = None
            if file_extension in file_type_folders:
                destination_folder = os.path.join(source_dir, file_type_folders[file_extension])

            # Create the destination folder if it doesn't exist
            if destination_folder and not os.path.exists(destination_folder):
                os.makedirs(destination_folder)

            # Move the file to the destination folder (handle potential errors)
            if destination_folder:
                try:
                    shutil.move(file_path, destination_folder)
                    print(f"Moved '{filename}' to '{destination_folder}'")
                except OSError as e:
                    print(f"Error moving '{filename}': {e}")

if __name__ == '__main__':
    organize_downloads(source_dir)
```

```python
import os
import shutil
import time  # For optional sleep after starting the observer

from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler


def organize_downloads(source_dir, filename):
    """
    Organizes a downloaded file based on its extension.

    Args:
        source_dir (str): The path to the downloads folder.
        filename (str): The name of the downloaded file.
    """

    # Define a dictionary to map file extensions to folder names (customize as needed)
    file_type_folders = {
        '.pdf': 'Documents/PDFs',
        '.docx': 'Documents/Word',
        '.xlsx': 'Documents/Excel',
        '.pptx': 'Documents/Presentations',
        '.jpg': 'Images',
        '.png': 'Images',
        '.gif': 'Images',
        '.mp3': 'Music',
        '.mp4': 'Videos',
        '.zip': 'Compressed',
        '.rar': 'Compressed',
    }

    # Construct the full file path
    file_path = os.path.join(source_dir, filename)

    if os.path.isfile(file_path):
        # Extract file extension (lowercase for case-insensitive matching)
        file_extension = os.path.splitext(filename)[1].lower()

        # Determine the destination folder based on the extension
        destination_folder = None
        if file_extension in file_type_folders:
            destination_folder = os.path.join(source_dir, file_type_folders[file_extension])
        else:
            # Use a default folder for unknown extensions
            destination_folder = os.path.join(source_dir, 'Uncategorized')

        # Create the destination folder if it doesn't exist
        if destination_folder and not os.path.exists(destination_folder):
            os.makedirs(destination_folder)

        # Move the file to the destination folder (handle potential errors)
        if destination_folder:
            try:
                shutil.move(file_path, destination_folder)
                print(f"Moved '{filename}' to '{destination_folder}'")
            except OSError as e:
                print(f"Error moving '{filename}': {e}")


class MyHandler(FileSystemEventHandler):
    def on_created(self, event):
        if event.is_file:  # Check if it's a file creation event
            # Get the filename from the event path
            filename = os.path.basename(event.src_path)
            organize_downloads(source_dir, filename)  # Trigger organization logic


if __name__ == "__main__":
    downloads_folder_path ="C:\\Users\\swati\\Downloads"
    observer = Observer()
    observer.schedule(MyHandler(), source_dir, recursive=False)  # Monitor only downloads folder

    # Optional: Allow a short time for observer initialization (uncomment if needed)
    # time.sleep(1)

    observer.start()
    try:
        while True:
            # Check for events more frequently (adjust timeout as needed)
            observer.join(timeout=0.2)
    except KeyboardInterrupt:
        observer.stop()
    observer.join()
```
